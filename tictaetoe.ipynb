# Tic Tac Toe

- Reglas del Juego "Tic-Tac-Toe":

    - El juego se juega en un tablero de 3x3.

    - Dos jugadores, uno representa "X" y el otro "O", alternan turnos.

    - El objetivo es obtener tres s√≠mbolos del mismo tipo en l√≠nea horizontal, vertical o diagonal.

    - El juego termina cuando un jugador gana o el tablero se llena (empate).

    - El jugador que logre colocar tres de sus s√≠mbolos en l√≠nea es el ganador.


- Pistas para las alumnas:

    1. Comienza creando una representaci√≥n del tablero de juego utilizando una lista de listas para mantener un seguimiento de las jugadas.

    2. Implementa una funci√≥n para imprimir el tablero en la pantalla en un formato claro y legible.

    3. Utiliza un bucle `while` para alternar los turnos de los jugadores y recopilar sus movimientos.

    4. Verifica despu√©s de cada movimiento si hay un ganador o si el juego termina en empate.

    5. Implementa una funci√≥n que verifique si un jugador ha ganado alineando tres s√≠mbolos en fila, columna o diagonal.

    6. Asegurate de validar que una casilla seleccionada por un jugador no est√© ocupada antes de permitir el movimiento.

```python
# --- Inicializar el tablero ---
tablero = [[" " for _ in range(3)] for _ in range(3)]

# --- Funci√≥n para imprimir el tablero ---
def imprimir_tablero():
    print("-------------")
    for fila in tablero:
        print("|", fila[0], "|", fila[1], "|", fila[2], "|")
        print("-------------")

# --- Funci√≥n para verificar si hay un ganador ---
def hay_ganador(jugador):
    # Revisar filas
    for i in range(3):
        if tablero[i][0] == jugador and tablero[i][1] == jugador and tablero[i][2] == jugador:
            return True

    # Revisar columnas
    for i in range(3):
        if tablero[0][i] == jugador and tablero[1][i] == jugador and tablero[2][i] == jugador:
            return True

    # Revisar diagonales
    if tablero[0][0] == jugador and tablero[1][1] == jugador and tablero[2][2] == jugador:
        return True
    if tablero[0][2] == jugador and tablero[1][1] == jugador and tablero[2][0] == jugador:
        return True

    return False

# --- Funci√≥n para saber si el tablero est√° lleno ---
def tablero_lleno():
    for fila in tablero:
        for celda in fila:
            if celda == " ":
                return False
    return True

# --- Bucle principal del juego ---
jugador_actual = "X"

while True:
    imprimir_tablero()
    print("Turno del jugador", jugador_actual)

    # Pedir jugada
    fila = int(input("Elige la fila (0, 1, 2): "))
    col = int(input("Elige la columna (0, 1, 2): "))

    # Validar jugada
    if fila < 0 or fila > 2 or col < 0 or col > 2:
        print("‚ö†Ô∏è Posici√≥n fuera del rango, intenta de nuevo.")
        continue
    if tablero[fila][col] != " ":
        print("‚ö†Ô∏è Esa casilla ya est√° ocupada, elige otra.")
        continue

    # Realizar movimiento
    tablero[fila][col] = jugador_actual

    # Verificar ganador
    if hay_ganador(jugador_actual):
        imprimir_tablero()
        print("üéâ ¬°Jugador", jugador_actual, "gana!")
        break

    # Verificar empate
    if tablero_lleno():
        imprimir_tablero()
        print("ü§ù ¬°Empate!")
        break

    # Cambiar de jugador
    if jugador_actual == "X":
        jugador_actual = "O"
    else:
        jugador_actual = "X"
```
