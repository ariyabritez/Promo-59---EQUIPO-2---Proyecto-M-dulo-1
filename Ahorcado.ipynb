*** LÃ³gica del juego (paso a paso)

1- Elegir una palabra secreta
Puede estar fija (por ejemplo "python") o elegirse al azar de una lista.

2- Mostrar la palabra oculta con guiones
Ejemplo: Si la palabra es python, se muestra _ _ _ _ _ _.

3- Pedir al jugador que ingrese una letra
A- Si la letra estÃ¡ en la palabra â†’ se revela en todas las posiciones correctas.
B- Si no estÃ¡, se cuenta como un error.

4- Contar los intentos fallidos
A- El jugador tiene un nÃºmero limitado de vidas (por ejemplo, 6).
B- Cada error resta una vida.

5- Comprobar si el jugador gana o pierde
A- Gana si adivina todas las letras.
B- Pierde si se queda sin vidas.

6- Mostrar el resultado final
Palabra completa y mensaje de victoria o derrota.

```python
import random

# 1. Lista de palabras y selecciÃ³n aleatoria
palabras = ["legos", "pelota", "barbie", "trompo", "plastilina"]
palabra = random.choice(palabras)

# 2. Preparar el tablero
letras = list(palabra)
adivinadas = ["_"] * len(palabra)

# 3. Variables de control
vidas = 6
letras_usadas = []

print("Â¡Bienvenido al juego del ahorcado!")
print("Adivina la palabra secreta:")
print(" ".join(adivinadas))

# 4. Bucle principal
while vidas > 0 and "_" in adivinadas:
    letra = input("Ingresa una letra: ").lower()

    # Verificar si la letra ya fue usada
    if letra in letras_usadas:
        print("Ya usaste esa letra. Intenta con otra.")
        continue
    letras_usadas.append(letra)

    # Verificar si la letra estÃ¡ en la palabra
    if letra in letras:
        for i in range(len(letras)):
            if letras[i] == letra:
                adivinadas[i] = letra
        print("Â¡Bien hecho! Letra correcta.")
    else:
        vidas -= 1
        print(f"Letra incorrecta. Te quedan {vidas} vidas.")
    
    # Mostrar el progreso actual
    print(" ".join(adivinadas))
    print("Letras usadas:", ", ".join(letras_usadas))

# 5. Resultado final
if "_" not in adivinadas:
    print("ðŸŽ‰ Â¡Ganaste! La palabra era:", palabra)
else:
    print("ðŸ’€ Â¡Perdiste! La palabra era:", palabra)
```
