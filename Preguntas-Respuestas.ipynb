# Preguntas y Respuestas (Sin Interfaz)

```python
import random

def jugar():
    preguntas = {
        "Â¿QuÃ© rÃ­o pasa por Budapest?": "Danubio",
        "Â¿CuÃ¡ntas comunidades autÃ³nomas tiene EspaÃ±a?": "17",
        "Â¿CuÃ¡l es la ciudad mÃ¡s poblada del mundo?": "Tokio",
        "Â¿CuÃ¡l es el rÃ­o mÃ¡s largo de EspaÃ±a?": "Tajo",
        "Â¿DÃ³nde estÃ¡ el Taj Mahal?": "India",
        "Â¿CuÃ¡ntas lenguas co-oficiales hay en EspaÃ±a?": "4",
        "Â¿En quÃ© continente estÃ¡ el monte Everest?": "Asia"
    }

    print("ğŸŒEL GRAN JUEGO DE PREGUNTAS Y RESPUESTAS DE GEOGRAFIA:ğŸŒ")


    # Pongo 8 preguntas porque damos lugar a que haya 5 aciertos + 3 errores antes de acabar el juego
    fallos = 0
    aciertos = 0

    while aciertos < 5 and fallos < 3:
        print(f"Aciertos: {aciertos}, Fallos: {fallos}")
        print("\n" * 1)  # Separador entre preguntas
        
        # random.choice posibilita que el cÃ³digo elija al azar una de las 8 preguntas
        pregunta, respuesta_correcta = random.choice(list(preguntas.items()))
        print(pregunta)
        
        respuesta_jugadora = input("Tu respuesta: ").strip().lower()
        print(f"Tu respuesta ha sido: {respuesta_jugadora}")

        # poner necesariamente .strip y .lower para eliminar errores de espacios y discriminaciÃ³n de mayÃºsculas/minÃºsculas
        if respuesta_jugadora == respuesta_correcta.strip().lower():
            aciertos += 1
            print("Â¡Â¡ğŸ‘ŒCorrecto!!")
        else:
            fallos += 1
            print("Â¡Â¡ğŸ™ˆHas fallado!!")
            print(f"La respuesta correcta era: {respuesta_correcta}")
        
        # .pop permite que dentro de random, una vez que haya salido una pregunta, quede eliminada y no se repita durante la partida
        preguntas.pop(pregunta)

    if aciertos == 5:
        print("\n" * 1)
        print("Â¡ğŸ¥³ğŸ‰Has ganado! Eres muy buena en geografÃ­a.")
    else:
        print("Â¡Â¡ğŸ˜©ğŸ‘Eliminada!! Otra vez serÃ¡...ğŸ¤·â€â™€ï¸")

while True:  # con la funcion de juego posibilitamos que al finalizar creemos un bucle para poder volver (o no) a jugar.
    jugar ()
    jugar_de_nuevo = input("Â¿Otra ronda de preguntas? (sÃ­/no): ").lower()
    if jugar_de_nuevo not in ["si", "sÃ­", "s"]:
        print("Â¡Hasta la prÃ³xima!")
        break
```

## Preguntas y Respuestas (Con Interfaz Tkinter)

```python
import tkinter as tk
from PIL import Image, ImageTk
import random
import os

class JuegoGeografia:
    def __init__(self, root):
        self.root = root
        self.root.title("Juego de GeografÃ­a")
        self.root.geometry("850x650")
        self.root.config(bg="#222831")

        # Frame para el logo y tÃ­tulo
        self.frame_inicio = tk.Frame(root, bg="#222831")
        self.frame_inicio.pack(pady=10)

        # Cargar el logo despuÃ©s de que la ventana se inicialice
        self.root.after(100, self.cargar_logo)

        # Texto de bienvenida
        tk.Label(root, text="Â¡HOLA ADALABER! \n\nBienvenida al juego de Preguntas Y Respuestas",
                 bg="#222831", fg="#FFD369", font=("Arial", 18, "bold")).pack(pady=10)

        # TÃ­tulo del juego
        tk.Label(root, text="ğŸŒ EL GRAN JUEGO DE GEOGRAFÃA: ğŸŒ",
                 bg="#222831", font=("Arial", 16, "bold")).pack(pady=10)

        # Diccionario de preguntas
        self.preguntas = {
            "Â¿QuÃ© rÃ­o pasa por Budapest?": "Danubio",
            "Â¿CuÃ¡ntas comunidades autÃ³nomas tiene EspaÃ±a?": "17",
            "Â¿CuÃ¡l es la ciudad mÃ¡s poblada del mundo?": "Tokio",
            "Â¿CuÃ¡l es el rÃ­o mÃ¡s largo de EspaÃ±a?": "Tajo",
            "Â¿DÃ³nde estÃ¡ el Taj Mahal?": "India",
            "Â¿CuÃ¡ntas lenguas co-oficiales hay en EspaÃ±a?": "4",
            "Â¿En quÃ© continente estÃ¡ el monte Everest?": "Asia"
        }
        self.reset_game()

        # Estado de aciertos y fallos
        self.label_estado = tk.Label(root, text="Aciertos: 0  Fallos: 0", font=("Arial", 13), bg="#222831")
        self.label_estado.pack(pady=4)

        # Pregunta actual
        tk.Label(root, text="Pregunta:", font=("Arial", 17), bg="#222831").pack(pady=4)
        self.label_pregunta = tk.Label(root, text="", font=("Arial", 14), bg="#222831")
        self.label_pregunta.pack(pady=8)

        # Entrada de respuesta
        self.entry_respuesta = tk.Entry(root, font=("Arial", 13))
        self.entry_respuesta.pack(pady=4)

        # BotÃ³n de enviar
        self.btn_enviar = tk.Button(root, text="Responder", font=("Arial", 12), command=self.comprobar_respuesta)
        self.btn_enviar.pack(pady=2)

        # Resultado
        self.label_resultado = tk.Label(root, text="", font=("Arial", 12))
        self.label_resultado.pack(pady=8)

        # BotÃ³n de nueva partida (oculto al inicio)
        self.btn_nueva_partida = tk.Button(root, text="Nueva partida", font=("Arial", 12), command=self.nueva_partida)

        # Inicia el juego con la primera pregunta
        self.nueva_pregunta()

    def cargar_logo(self):
        ruta_logo = os.path.join("assets", "logo-adalab.png")
        if os.path.exists(ruta_logo):
            imagen = Image.open(ruta_logo)
            imagen = imagen.resize((150, 150), Image.Resampling.LANCZOS)
            self.foto_logo = ImageTk.PhotoImage(imagen)
            self.label_logo = tk.Label(self.frame_inicio, image=self.foto_logo, bg="#222831")
            self.label_logo.pack(pady=10)
        else:
            tk.Label(self.frame_inicio, text="(Logo no encontrado)", bg="#222831", fg="#FFD369").pack(pady=10)

    def reset_game(self):
        self.preguntas_partida = self.preguntas.copy()
        self.aciertos = 0
        self.fallos = 0

    def nueva_partida(self):
        self.btn_nueva_partida.pack_forget()
        self.reset_game()
        self.label_resultado.config(text="")
        self.label_estado.config(text=f"Aciertos: {self.aciertos}  Fallos: {self.fallos}")
        self.btn_enviar.config(state="normal")
        self.entry_respuesta.config(state="normal")
        self.nueva_pregunta()

    def nueva_pregunta(self):
        if not self.preguntas_partida:
            self.finalizar_juego()
            return
        self.pregunta, self.respuesta = random.choice(list(self.preguntas_partida.items()))
        self.label_pregunta.config(text=self.pregunta)
        self.entry_respuesta.config(state="normal")
        self.entry_respuesta.delete(0, tk.END)
        self.btn_enviar.config(state="normal")

    def comprobar_respuesta(self):
        respuesta_jugadora = self.entry_respuesta.get().strip().lower()
        correcta = self.respuesta.strip().lower()
        if respuesta_jugadora == correcta:
            self.aciertos += 1
            self.label_resultado.config(text="Â¡Â¡Correcto ğŸ‘Œ !!", bg="#222831", font=("Arial", 16))
        else:
            self.fallos += 1
            self.label_resultado.config(
                text=f"Â¡Â¡Has fallado ğŸ™ˆ !! \nLa respuesta correcta era:\n {self.respuesta}",
                bg="#222831", font=("Arial", 16)
            )

        self.label_estado.config(text=f"Aciertos: {self.aciertos}  Fallos: {self.fallos}")
        self.preguntas_partida.pop(self.pregunta)

        if self.aciertos == 5 or self.fallos == 3:
            self.finalizar_juego()
        else:
            self.nueva_pregunta()

    def finalizar_juego(self):
        self.label_pregunta.config(text="")
        self.entry_respuesta.delete(0, tk.END)
        self.entry_respuesta.config(state="disabled")
        self.btn_enviar.config(state="disabled")
        if self.aciertos == 5:
            self.label_resultado.config(text="Â¡Has ganado! Eres muy buena en geografÃ­a. ğŸ¥³ğŸ‰",
                                       bg="#222831", font=("Arial", 16))
        else:
            self.label_resultado.config(text="Â¡Â¡Eliminada!! Otra vez serÃ¡...ğŸ¤·â€â™€ï¸",
                                       bg="#222831", font=("Arial", 16))
        print("DEBUG: Mostrando botÃ³n 'Nueva partida'")
        self.btn_nueva_partida.pack(pady=10)
        self.root.update_idletasks()


# Ejecutar interfaz
if __name__ == "__main__":
    root = tk.Tk()
    app = JuegoGeografia(root)
    root.mainloop()
```
