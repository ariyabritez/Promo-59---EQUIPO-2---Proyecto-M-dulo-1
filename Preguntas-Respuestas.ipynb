# Preguntas y Respuestas

- Reglas del Juego de "Preguntas y Respuestas de Geograf√≠a":
    - El juego consiste en hacer preguntas de geograf√≠a y que el jugador adivine la respuesta correcta.
    - Cada pregunta tiene una respuesta √∫nica y correcta.
    - El jugador tiene un n√∫mero limitado de intentos para responder correctamente a cada pregunta.
    - El juego sigue hasta que el jugador responda incorrectamente 3 de veces o haya respondido 5 preguntas correctamente.


- Pistas para las alumnas:
    1. Comienza creando un diccionario que contenga preguntas de geograf√≠a como claves y respuestas como valores.
    2. Implementa una funci√≥n para seleccionar una pregunta aleatoria del diccionario.
    3. Utiliza un bucle `while` para hacer preguntas de geograf√≠a hasta que el jugador responda un n√∫mero espec√≠fico de preguntas correctamente o alcance un n√∫mero m√°ximo de respuestas incorrectas.
    4. Dentro del bucle, presenta la pregunta al jugador y permita que ingresen sus respuestas.
    5. Lleva un seguimiento del n√∫mero de preguntas realizadas y el n√∫mero de respuestas correctas.
    6. Determina las condiciones de victoria y derrota del juego y sal del bucle cuando se cumplan estas condiciones.

1..Para elegir una pregunta al azar del diccionario...importar RANDOM

3...contador inicial de errores y aciertos a 0...BUCLE WHILE: Si <5 aciertos gana si <3 errores pierde
4...Mostrar la pregunta a la jugadora + Recibir su respuesta (input) + Comprobar si acierta o falla
5. Actualizar los contadores (aciertos y fallos)
 Las preguntas no se pueden repetir por eso ponemos un .pop
"""

# Para ejecutar el juego en una interface lo guardamos como un archivo de Python e importamos la libreria de Tkinter ------- preguntas-respuestas_tk_interface.py
# En la terminal del juego ponemos 'python preguntas-respuestas_tk_interface.py' para ejecutarlo

```python
import random

def jugar():
    preguntas = {
        "¬øQu√© r√≠o pasa por Budapest?": "Danubio",
        "¬øCu√°ntas comunidades aut√≥nomas tiene Espa√±a?": "17",
        "¬øCu√°l es la ciudad m√°s poblada del mundo?": "Tokio",
        "¬øCu√°l es el r√≠o m√°s largo de Espa√±a?": "Tajo",
        "¬øD√≥nde est√° el Taj Mahal?": "India",
        "¬øCu√°ntas lenguas co-oficiales hay en Espa√±a?": "4",
        "¬øEn qu√© continente est√° el monte Everest?": "Asia"
    }

    print("üåçEL GRAN JUEGO DE PREGUNTAS Y RESPUESTAS DE GEOGRAFIA:üåç")


    # Pongo 8 preguntas porque damos lugar a que haya 5 aciertos + 3 errores antes de acabar el juego
    fallos = 0
    aciertos = 0

    while aciertos < 5 and fallos < 3:
        print(f"Aciertos: {aciertos}, Fallos: {fallos}")
        print("\n" * 1)  # Separador entre preguntas
        
        # random.choice posibilita que el c√≥digo elija al azar una de las 8 preguntas
        pregunta, respuesta_correcta = random.choice(list(preguntas.items()))
        print(pregunta)
        
        respuesta_jugadora = input("Tu respuesta: ").strip().lower()
        print(f"Tu respuesta ha sido: {respuesta_jugadora}")

        # poner necesariamente .strip y .lower para eliminar errores de espacios y discriminaci√≥n de may√∫sculas/min√∫sculas
        if respuesta_jugadora == respuesta_correcta.strip().lower():
            aciertos += 1
            print("¬°¬°üëåCorrecto!!")
        else:
            fallos += 1
            print("¬°¬°üôàHas fallado!!")
            print(f"La respuesta correcta era: {respuesta_correcta}")
        
        # .pop permite que dentro de random, una vez que haya salido una pregunta, quede eliminada y no se repita durante la partida
        preguntas.pop(pregunta)

    if aciertos == 5:
        print("\n" * 1)
        print("¬°ü•≥üéâHas ganado! Eres muy buena en geograf√≠a.")
    else:
        print("¬°¬°üò©üëéEliminada!! Otra vez ser√°...ü§∑‚Äç‚ôÄÔ∏è")

while True:  # con la funcion de juego posibilitamos que al finalizar creemos un bucle para poder volver (o no) a jugar.
    jugar ()
    jugar_de_nuevo = input("¬øOtra ronda de preguntas? (s√≠/no): ").lower()
    if jugar_de_nuevo not in ["si", "s√≠", "s"]:
        print("¬°Hasta la pr√≥xima!")
        break
```

```python
import tkinter as tk
from PIL import Image, ImageTk
import random
import os

class JuegoGeografia:
    def __init__(self, root):
        self.root = root
        self.root.title("Juego de Geograf√≠a")
        self.root.geometry("850x650")
        self.root.config(bg="#222831")
        
        # Frame para el logo y t√≠tulo
        frame_inicio = tk.Frame(root, bg="#222831")
        frame_inicio.pack(pady=10)
        
        
        # Logo Adalab
        ruta_logo = os.path.join("assets", "logo-adalab.png")
        
        if os.path.exists(ruta_logo):
            imagen = Image.open(ruta_logo)
            imagen = imagen.resize((150, 150), Image.Resampling.LANCZOS)
            self.logo_global = ImageTk.PhotoImage(imagen)
            tk.Label(frame_inicio, image=self.logo_global, bg="#222831").pack(pady=10)
        else:
            tk.Label(frame_inicio, text="(Logo no encontrado)", bg="#222831", fg="#FFD369").pack(pady=10)
        
        frase_adalab = tk.Label(root, text="¬°HOLA ADALABER! \n\nBienvenida al juego de Preguntas Y Respuestas", bg="#222831", fg="#FFD369", font=("Arial", 18, "bold")).pack(pady=10)
        titulo = tk.Label(root, text="üåç EL GRAN JUEGO DE GEOGRAF√çA: üåç", bg="#222831", font=("Arial", 16, "bold"))
        titulo.pack(pady=10)

        self.preguntas = {
            "¬øQu√© r√≠o pasa por Budapest?": "Danubio",
            "¬øCu√°ntas comunidades aut√≥nomas tiene Espa√±a?": "17",
            "¬øCu√°l es la ciudad m√°s poblada del mundo?": "Tokio",
            "¬øCu√°l es el r√≠o m√°s largo de Espa√±a?": "Tajo",
            "¬øD√≥nde est√° el Taj Mahal?": "India",
            "¬øCu√°ntas lenguas co-oficiales hay en Espa√±a?": "4",
            "¬øEn qu√© continente est√° el monte Everest?": "Asia"
        }
        self.reset_game()
        
        self.label_estado = tk.Label(root, text="Aciertos: 0  Fallos: 0", font=("Arial", 13),bg="#222831")
        self.label_estado.pack(pady=4)
        
        self.label_titulo_pregunta = tk.Label(root, text="Pregunta:", font=("Arial", 17),bg="#222831")
        self.label_estado.pack(pady=4)
        
        self.label_pregunta = tk.Label(root, text="", font=("Arial",14), bg="#222831")
        self.label_pregunta.pack(pady=8)
        
        self.entry_respuesta = tk.Entry(root, font=("Arial", 13))
        self.entry_respuesta.pack(pady=4)
        
        self.btn_enviar = tk.Button(root, text="Responder", font=("Arial", 12), command=self.comprobar_respuesta)
        self.btn_enviar.pack(pady=2)

        self.label_resultado = tk.Label(root, text="", font=("Arial",12))
        self.label_resultado.pack(pady=8)

        self.btn_nueva_partida = tk.Button(root, text="Nueva partida", font=("Arial", 12), command=self.nueva_partida)
        # NO pack aqu√≠

        self.nueva_pregunta()

    def reset_game(self):
        self.preguntas_partida = self.preguntas.copy()
        self.aciertos = 0
        self.fallos = 0

    def nueva_partida(self):
        # FUERZA que se oculte el bot√≥n
        self.btn_nueva_partida.pack_forget()
        self.reset_game()
        self.label_resultado.config(text="")
        self.label_estado.config(text=f"Aciertos: {self.aciertos}  Fallos: {self.fallos}")
        self.btn_enviar.config(state="normal")
        self.entry_respuesta.config(state="normal")
        self.nueva_pregunta()

    def nueva_pregunta(self):
        if not self.preguntas_partida:
            self.finalizar_juego()
            return
        self.pregunta, self.respuesta = random.choice(list(self.preguntas_partida.items()))
        self.label_pregunta.config(text=self.pregunta)
        self.entry_respuesta.config(state="normal")
        self.entry_respuesta.delete(0, tk.END)
        self.btn_enviar.config(state="normal")

    def comprobar_respuesta(self):
        respuesta_jugadora = self.entry_respuesta.get().strip().lower()
        correcta = self.respuesta.strip().lower()
        if respuesta_jugadora == correcta:
            self.aciertos += 1
            self.label_resultado.config(text="¬°¬°Correcto üëå !!", bg="#222831", font=("Arial",16))
        else:
            self.fallos += 1
            self.label_resultado.config(text=f"¬°¬°Has fallado üôà !! \nLa respuesta correcta era:\n {self.respuesta}", bg="#222831", font=("Arial",16))

        self.label_estado.config(text=f"Aciertos: {self.aciertos}  Fallos: {self.fallos}")
        self.preguntas_partida.pop(self.pregunta)

        if self.aciertos == 5 or self.fallos == 3:
            self.finalizar_juego()
        else:
            self.nueva_pregunta()

    def finalizar_juego(self):
        self.label_pregunta.config(text="")
        self.entry_respuesta.delete(0, tk.END)
        self.entry_respuesta.config(state="disabled")
        self.btn_enviar.config(state="disabled")
        if self.aciertos == 5:
            self.label_resultado.config(text="¬°Has ganado! Eres muy buena en geograf√≠a. ü•≥üéâ", bg="#222831",font=("Arial",16))
        else:
            self.label_resultado.config(text="¬°¬°Eliminada!! Otra vez ser√°...ü§∑‚Äç‚ôÄÔ∏è", bg="#222831", font=("Arial",16))
        print("DEBUG: Mostrando bot√≥n 'Nueva partida'")  # L√≠nea para ver salida en consola
        self.btn_nueva_partida.pack(pady=10)
        self.root.update_idletasks()  # Fuerza update visual

# Ejecutar interfaz
if __name__ == "__main__":
    root = tk.Tk()
    app = JuegoGeografia(root)
    root.mainloop()
```
